```python
import re
word_dict = dict()
file = open("LICENSE.txt")
for line in file:
    for word in re.findall(r'\w+', line):
        if word in '0123456789':
            continue
            
        if word in word_dict.keys():
            word_dict[word] += 1
        else:
            word_dict[word] = 1
            
sorted(word_dict, key=word_dict.__getitem__, reverse=True)[:11]            
```
