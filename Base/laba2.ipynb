```python
# Создадим пустой словать Capitals
Capitals = dict()

# Заполним его несколькими значениями
Capitals['Russia'] = 'Moscow'
Capitals['Ukraine'] = 'Kiev'
Capitals['USA'] = 'Washington'

# Считаем название страны
print('В какой стране вы живете?')
country = input()

# Проверим, есть ли такая страна в словаре Capitals
if country in Capitals:
    # Если есть - выведем ее столицу
    print('Столица вашей страны', Capitals[country])
else:
    # Запросим название столицы и добавим его в словарь
    print('Как называется столица вашей страны?')
    city = input()
    Capitals[country] = city
    
print(Capitals)
```

```python
Country = {}
c = open('count.txt', encoding='utf-8')
for line in c:
    coun, lan = line.split(' : ')
    lan = str(lan.split())
    Country[coun] = lan
Language = {}

for key in Country:
    for val in Country[key]:
        if val not in Language:
            Language[val] = []
        Language[val].append(key)
#print(Language)

```

```python
with open('count.txt', encoding='utf-8') as country_lang:
    lang_country = {}
    for line in country_lang:
        country, languages = line.split(' : ')
        languages = languages.strip().split()
        for lang in languages:
            if lang not in lang_country:
                lang_country[lang] = []
            lang_country[lang].append(country)
            
#print(lang_country)        
```

```python
enru = {}
ruen_all = {}
enru_all = {}

eng = open('enru.txt', encoding='utf-8')
for line in eng:
    word, tr = line.split('\t-\t')
    tri = tr.split()
    for x in tri:
        tr = x
        enru[word] = tr
        

        
ruen = {}
rus = open('ruen.txt', encoding='utf-8')
for line in rus:
    word, tr = line.split('\t-\t')
    tri = tr.split()
    for x in tri:
        tr = x
        ruen[word] = tr
        

ruen2 = {v:k for k, v in enru.items()}
for key in ruen:
    if key in ruen2:
        ruen_all[key] = [d[key][0] for d in (ruen, ruen2)]
ruen_all.update(ruen)
ruen_all.update(ruen2)


enru2 = {v:k for k, v in ruen.items()}
for key in enru:
    if key in enru2:
        enru_all[key] = [d[key][0] for d in (enru, enru2)]      
enru_all.update(enru)
enru_all.update(enru2)
```
