# Упражнение №4
Фигура Лиссажу задаётся выражением: x(t, a) = sin(t + a), y(t) = cos(2*t)

Используя matplotlib анимируйте фигуру Лиссажу, меняя в каждом кадре значение параметра a.

```python
import pylab
import math
from matplotlib import mlab

t_list = mlab.frange(-20, 20, 0.01)

pylab.ion()

for n in range(50):
    y_list = [math.cos(2 * t) for t in t_list]
    x_list = [math.sin(t + n) for t in t_list]
    pylab.clf()
    pylab.plot(x_list, y_list)
    pylab.draw()
    pylab.pause(0.1)


pylab.close()
```

Приблизить данные из приведённого примера с погрешностями или свои собственные (из лабораторного практикума по общей физике) многочленами первой и второй степени. Начертить точки с погрешностями и полученные аппроксимационные кривые на одном графике.

```python
x = [[1, 2, 3, 4, 5]]
y = [0.99, 0.49, 0.35, 0.253, 0.18]
plt.errorbar(x, y, xerr=0.05, yerr=0.1)
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 1.42, 1.76, 2, 2.24, 2.5]
p1 = np.polyfit(x, y, deg=1)
p2 = np.polyfit(x, y, deg=2)
pf_1 = np.poly1d(p1)
pf_2 = np.poly1d(p2)
A = []
B = []
for i in x:
    A.append(pf_1(i))
    B.append(pf_2(i))
plt.plot(x, A, x, B)
plt.errorbar(x, y, xerr=0.05, yerr=0.1)
plt.show()
```
